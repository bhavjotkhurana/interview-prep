Learning linear regression using some code

```python
import pandas as pd
import numpy as np 

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

import statsmodels.api as sm

%matplotlib inline
```

```python
data = pd.read_csv('~/Desktop/interview-prep/data-science/Code/Advertising.csv', index_col = 0)
data.head()
```

# Simple Linear Regression

```python
plt.figure(figsize=(16,8))
plt.scatter(data['TV'], data['Sales'], color='black')
plt.xlabel('TV Advertising Spend (in thousands of dollars)')
plt.ylabel('Sales (in thousands of units)')
plt.show()
```

```python
X = data['TV'].values.reshape(-1,1)
y = data['Sales'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(X, y)

print(f"The linear model is: \n Y = {reg.intercept_[0]} + {reg.coef_[0][0]} * TV")
```

```python
predictions = reg.predict(X)

plt.figure(figsize = (16,8))
plt.scatter(X, y, c = 'black')
plt.plot(X, predictions, c = 'blue', linewidth = 2)
plt.xlabel('TV Advertising Spend (in thousands of dollars)')
plt.ylabel('Sales (in thousands of units)')
plt.show()
```

```python
X = data['TV']
y = data['Sales']

exog = sm.add_constant(X)
est = sm.OLS(y, exog).fit()

print(est.summary())
```

## Multiple Linear Regression

```python
Xs = data.drop(['Sales'], axis = 1)
y = data['Sales'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(Xs, y)

print(f"The multiple linear regression model is: \n Y = {reg.intercept_[0]} + {reg.coef_[0][0]} * TV + {reg.coef_[0][1]} * Radio + {reg.coef_[0][2]} * Newspaper")
```

```python
X = np.column_stack((data['TV'], data['Radio'], data['Newspaper']))
y = data['Sales'].values.reshape(-1,1)

exog = sm.add_constant(X)
est = sm.OLS(y, exog).fit()

print(est.summary())
```
