# **404. Sum of Left Leaves**  
**Difficulty:** Easy

---

**Problem Statement**

Given the root of a binary tree, return the sum of all left leaves.

A **leaf** is a node with no children.  
A **left leaf** is a leaf that is the left child of another node.

---

**Example 1**

Input:  
`root = [3,9,20,null,null,15,7]`

Output:  
`24`

Explanation:  
There are two left leaves in the binary tree, with values 9 and 15 respectively.

---

**Example 2**

Input:  
`root = [1]`

Output:  
`0`

---

**Constraints**

- The number of nodes in the tree is in the range `[1, 1000]`
- `-1000 <= Node.val <= 1000`


```python
# %%
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values):
    if not values:
        return None

    nodes = [None if v is None else TreeNode(v) for v in values]
    kids = nodes[::-1]
    root = kids.pop()

    for node in nodes:
        if node:
            if kids:
                node.left = kids.pop()
            if kids:
                node.right = kids.pop()

    return root


class Solution:
    def sumOfLeftLeaves(self, root: TreeNode) -> int:
        total = 0
        def leaves(node, isLeft): 
            nonlocal total
            if not node: 
                return
            if isLeft and not node.left and not node.right: 
                total += node.val
                return
            leaves(node.left, True)
            leaves(node.right, False)

        leaves(root, False)
        return total
```

```python
# %%
# Test scaffold

solution = Solution()

# Example 1
root1 = build_tree([3, 9, 20, None, None, 15, 7])
print(solution.sumOfLeftLeaves(root1))
# Expected output: 24

# Example 2
root2 = build_tree([1])
print(solution.sumOfLeftLeaves(root2))
# Expected output: 0
```
