# 1. Palindrome String

**Last Updated:** 23 Jul, 2025

Given a string `s`, the task is to check if it is palindrome or not.

---

## Examples

### Example 1
**Input:**  
s = `"abba"`  

**Output:**  
1  

**Explanation:**  
`s` is a palindrome.

---

### Example 2
**Input:**  
s = `"abc"`  

**Output:**  
0  

**Explanation:**  
`s` is not a palindrome.


```python
class Solution:
    def isPalindrome(self, s: str) -> int:
        left = 0 
        right = len(s) - 1
        while left < right: 
            if s[left] == s[right]: 
                left += 1
                right -= 1
            else: 
                return 0
        return 1
```

```python
# Test scaffold

sol = Solution()

# Example 1
s1 = "abba"
print(sol.isPalindrome(s1))  # Expected output: 1

# Example 2
s2 = "abc"
print(sol.isPalindrome(s2))  # Expected output: 0
```

# 2. Reverse a String

**Last Updated:** 3 Oct, 2025

Given a string `s`, the task is to **reverse the string**.  
Reversing a string means rearranging the characters such that the **first** character becomes the **last**, the **second** character becomes **second last**, and so on.

---

## Examples

### Example 1
**Input:**  
s = `"GeeksforGeeks"`

**Output:**  
`"skeeGrofskeeG"`

**Explanation:**  
The first character `G` moves to the last position, the second character `e` moves to second-last, and so on.

---

### Example 2
**Input:**  
s = `"abdcfe"`

**Output:**  
`"efcdba"`

**Explanation:**  
The first character `a` moves to the last position, the second character `b` moves to second-last, and so on.


```python
class Solution:
    def reverseString(self, s: str) -> str:
        arr = list(s)
        left = 0 
        right = len(arr) - 1
        while left < right: 
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
        return "".join(arr)
```

```python
# Test scaffold

sol = Solution()

# Example 1
s1 = "GeeksforGeeks"
print(sol.reverseString(s1))  # Expected output: "skeeGrofskeeG"

# Example 2
s2 = "abdcfe"
print(sol.reverseString(s2))  # Expected output: "efcdba"
```

# 3. Reverse words in a string

**Last Updated:** 29 Oct, 2025

Given a string `s`, find a way to reverse the order of the words in the given string.

**Note:**  
The string may contain leading or trailing dots (`.`) or multiple dots between two words.  
The returned string should only have a **single dot (`.`)** separating the words.

---

## Examples

### Example 1
**Input:**  
s = `"i.like.this.program.very.much"`

**Output:**  
`"much.very.program.this.like.i"`

**Explanation:**  
The words in the input string are reversed while maintaining the dots as separators, resulting in `"much.very.program.this.like.i"`.

---

### Example 2
**Input:**  
s = `"..geeks..for.geeks."`

**Output:**  
`"geeks.for.geeks"`

---

### Example 3
**Input:**  
s = `"...home....."`

**Output:**  
`"home"`


```python
class Solution:
    def reverseWords(self, s: str) -> str:
        arr = s.split(".")
        res = []
        for i in range(len(arr) - 1, -1, -1): 
            if arr[i] != '': 
                res.append(arr[i])
        return ".".join(res)
```

```python
# Test scaffold

sol = Solution()

# Example 1
s1 = "i.like.this.program.very.much"
print(sol.reverseWords(s1))  # Expected output: "much.very.program.this.like.i"

# Example 2
s2 = "..geeks..for.geeks."
print(sol.reverseWords(s2))  # Expected output: "geeks.for.geeks"

# Example 3
s3 = "...home....."
print(sol.reverseWords(s3))  # Expected output: "home"
```

# 4. Check if Strings Are Rotations of Each Other

**Last Updated:** 3 Oct, 2025

Given two strings `s1` and `s2` of equal length, determine whether `s2` is a rotation of `s1`.

A string is said to be a rotation of another if it can be obtained by shifting some leading characters of the original string to its end without changing the order of characters.

---

## Examples

### Example 1
**Input:**  
s1 = `"abcd"`, s2 = `"cdab"`

**Output:**  
true

**Explanation:**  
After 2 right rotations, `s1` will become equal to `s2`.

---

### Example 2
**Input:**  
s1 = `"aab"`, s2 = `"aba"`

**Output:**  
true

**Explanation:**  
After 1 left rotation, `s1` will become equal to `s2`.

---

### Example 3
**Input:**  
s1 = `"abcd"`, s2 = `"acbd"`

**Output:**  
false

**Explanation:**  
Strings are not rotations of each other.


```python
class Solution:
    def areRotations(self, s1: str, s2: str) -> bool:
        if len(s1) != len(s2): 
            return False
        find = s1[0]
        idx = 0
        f = -1
        for i in range(len(s2)): 
            if s2[i] == find: 
                f = i
                break
   
        for i in range(f, len(s2)): 
            if s1[idx] != s2[i]: 
                return False
            idx += 1

        for i in range(0, f): 
            if s1[idx] != s2[i]: 
                return False
            idx += 1
        return True
```

```python
class Solution:
    def areRotations(self, s1: str, s2: str) -> bool:
        s = s1 + s1
        return s2 in s
```

```python
# Test scaffold

sol = Solution()

# Example 1
s1, s2 = "abcd", "cdab"
print(sol.areRotations(s1, s2))  # Expected output: True

# Example 2
s1, s2 = "aab", "aba"
print(sol.areRotations(s1, s2))  # Expected output: True

# Example 3
s1, s2 = "abcd", "acbd"
print(sol.areRotations(s1, s2))  # Expected output: False
```

# 5. First non-repeating character of given string

**Last Updated:** 7 Apr, 2025

Given a string `s` of lowercase English letters, the task is to find the **first non-repeating character**.  
If there is no such character, return `'$'`.

---

## Examples

### Example 1
**Input:**  
s = `"geeksforgeeks"`

**Output:**  
`'f'`

**Explanation:**  
`'f'` is the first character in the string which does not repeat.

---

### Example 2
**Input:**  
s = `"racecar"`

**Output:**  
`'e'`

**Explanation:**  
`'e'` is the only character in the string which does not repeat.

---

### Example 3
**Input:**  
s = `"aabbccc"`

**Output:**  
`'$'`

**Explanation:**  
All the characters in the given string are repeating.


```python
class Solution:
    def firstNonRepeating(self, s: str) -> str:
        store = {}
        f = '$'
        for c in s: 
            if c in store: 
                store[c] += 1
            else: 
                store[c] = 1
        for key, value in store.items(): 
            if store[key] == 1: 
                f = key
                break
        return f
```

```python
# Test scaffold

sol = Solution()

# Example 1
s1 = "geeksforgeeks"
print(sol.firstNonRepeating(s1))  # Expected output: 'f'

# Example 2
s2 = "racecar"
print(sol.firstNonRepeating(s2))  # Expected output: 'e'

# Example 3
s3 = "aabbccc"
print(sol.firstNonRepeating(s3))  # Expected output: '$'
```

# 6. String to Integer – Write your own atoi()

**Last Updated:** 4 Aug, 2025

Given a string `s`, convert it into integer format **without utilizing any built-in functions**.  
Refer the below steps to know about the `atoi()` function.

---

## Rules

- Skip any leading whitespaces.
- Check for a sign (`'+'` or `'-'`), default to positive if no sign is present.
- Read the integer by ignoring leading zeros until a non-digit character is encountered or the end of the string is reached.  
  If no digits are present, return `0`.
- If the integer is greater than `2^31 - 1`, return `2^31 - 1`.
- If the integer is smaller than `-2^31`, return `-2^31`.

---

## Examples

### Example 1
**Input:**  
s = `"-123"`

**Output:**  
`-123`

---

### Example 2
**Input:**  
s = `"   "`

**Output:**  
`0`

**Explanation:**  
No digits are present, therefore return `0`.

---

### Example 3
**Input:**  
s = `"1231231231311133"`

**Output:**  
`2147483647`

**Explanation:**  
The converted number is greater than `2^31 - 1`, therefore return `2147483647`.

---

### Example 4
**Input:**  
s = `"-999999999999"`

**Output:**  
`-2147483648`

**Explanation:**  
The converted number is smaller than `-2^31`, therefore return `-2147483648`.

---

### Example 5
**Input:**  
s = `"-0012gfg4"`

**Output:**  
`-12`

**Explanation:**  
Nothing is read after `-12` as a non-digit character `'g'` was encountered.


```python
class Solution:
    def myAtoi(self, s: str) -> int:
        INT_MIN = -(2**31)
        INT_MAX = 2**31 - 1

        n = len(s)
        idx = 0

        # 1) Skip leading whitespace
        while idx < n and s[idx] == " ":
            idx += 1

        # 2) Optional sign
        sign = 1
        if idx < n and s[idx] in "+-":
            if s[idx] == "-":
                sign = -1
            idx += 1

        # 3) Read digits
        num = 0
        while idx < n and "0" <= s[idx] <= "9":
            digit = ord(s[idx]) - ord("0")
            num = num * 10 + digit
            idx += 1

        num *= sign

        # 4) Clamp to 32-bit signed range
        if num < INT_MIN:
            return INT_MIN
        if num > INT_MAX:
            return INT_MAX
        return num
```

```python
# Test scaffold

sol = Solution()

# Example 1
print(sol.myAtoi("-123"))  # Expected output: -123

# Example 2
print(sol.myAtoi("   "))  # Expected output: 0

# Example 3
print(sol.myAtoi("1231231231311133"))  # Expected output: 2147483647

# Example 4
print(sol.myAtoi("-999999999999"))  # Expected output: -2147483648

# Example 5
print(sol.myAtoi("-0012gfg4"))  # Expected output: -12
```

# Encrypt the string - 2

**Last Updated:** 23 Jul, 2025

Given a lowercase English string `s` of length `n`, encrypt it as follows:

- Replace each **contiguous group of the same character** with the character followed by the **lowercase hexadecimal representation** of its length.
- Reverse the entire modified string.

Return the encrypted string.

**Note:**  
All hexadecimal letters should be converted to **lowercase** letters.

---

## Examples

### Example 1
**Input:**  
s = `"aaaaaaaaaaa"`

**Output:**  
`"ba"`

**Explanation:**  
- Count consecutive characters: `"aaaaaaaaaaa"` → `"a11"`  
- Convert frequency to hexadecimal: `"11"` → `"b"`  
- Reverse the string: `"ab"` → `"ba"`

---

### Example 2
**Input:**  
s = `"abc"`

**Output:**  
`"1c1b1a"`

**Explanation:**  
- Convert to character-frequency format: `"abc"` → `"a1b1c1"`  
- Convert counts to hexadecimal (no change as `1` remains `1`)  
- Reverse the string: `"a1b1c1"` → `"1c1b1a"`


```python
class Solution:
    def encryptString(self, s: str) -> str:
        if not s: 
            return ""
        enc = ""
        prev = s[0]
        count = 1
        for i in range(1, len(s)): 
            curr = s[i]
            if prev == curr: 
                count += 1
            else: 
                enc += prev + hex(count)[2:]
                count = 1
                prev = curr
        
        enc += prev + hex(count)[2:]

        return enc[::-1]
```

```python
# Test scaffold

sol = Solution()

# Example 1
s1 = "aaaaaaaaaaa"
print(sol.encryptString(s1))  # Expected output: "ba"

# Example 2
s2 = "abc"
print(sol.encryptString(s2))  # Expected output: "1c1b1a"
```

# Find the Missing Number
**Last Updated:** 19 Apr, 2025

Given an array `arr[]` of size `n-1` with **distinct integers** in the range of `[1, n]`. This array represents a permutation of the integers from `1` to `n` with one element missing. Find the missing element in the array.

## Examples

**Input:** `arr[] = [8, 2, 4, 5, 3, 7, 1]`  
**Output:** `6`  
**Explanation:** All the numbers from `1` to `8` are present except `6`.

**Input:** `arr[] = [1, 2, 3, 5]`  
**Output:** `4`  
**Explanation:** Here the size of the array is `4`, so the range will be `[1, 5]`. The missing number between `1` to `5` is `4`.


```python
from typing import List

class Solution:
    def missingNumber(self, arr: List[int]) -> int:
        n = len(arr) + 1
        sum = (n * (n + 1)) / 2
        for a in arr: 
            sum -= a
        return int(sum)
```

```python
# Test scaffold (from examples)

sol = Solution()

arr1 = [8, 2, 4, 5, 3, 7, 1]
print(sol.missingNumber(arr1))
# Expected: 6

arr2 = [1, 2, 3, 5]
print(sol.missingNumber(arr2))
# Expected: 4
```

# Second Largest Element in an Array
**Last Updated:** 10 Feb, 2025

Given an array of **positive integers** `arr[]` of size `n`, the task is to find the **second largest distinct element** in the array.

**Note:** If the second largest element does not exist, return `-1`.

## Examples

**Input:** `arr[] = [12, 35, 1, 10, 34, 1]`  
**Output:** `34`  
**Explanation:** The largest element of the array is `35` and the second largest element is `34`.

**Input:** `arr[] = [10, 5, 10]`  
**Output:** `5`  
**Explanation:** The largest element of the array is `10` and the second largest element is `5`.

**Input:** `arr[] = [10, 10, 10]`  
**Output:** `-1`  
**Explanation:** The largest element of the array is `10`; there is no second largest element.


```python
from typing import List

class Solution:
    def secondLargest(self, arr: List[int]) -> int:
        first = -1
        second = -1
        for n in arr: 
            if n > first: 
                second = first
                first = n
            elif n > second and n != first: 
                second = n
        return second
```

```python
# Test scaffold (from examples)

sol = Solution()

arr1 = [12, 35, 1, 10, 34, 1]
print(sol.secondLargest(arr1))
# Expected: 34

arr2 = [10, 5, 10]
print(sol.secondLargest(arr2))
# Expected: 5

arr3 = [10, 10, 10]
print(sol.secondLargest(arr3))
# Expected: -1
```

# Transition point in a binary array
**Last Updated:** 19 Sep, 2025

Given a **sorted array** `arr[]`, containing only `0`s and `1`s, find the **transition point**, i.e., the **first index where `1` was observed**, and before that, only `0` was observed.

- If `arr` does not have any `1`, return `-1`.
- If the array does not have any `0`, return `0`.

## Examples

**Input:** `0 0 0 1 1`  
**Output:** `3`  
**Explanation:** Index `3` is the transition point where first `1` was observed.

**Input:** `arr[] = [0, 0, 0, 0]`  
**Output:** `-1`  
**Explanation:** Since there is no `1`, the answer is `-1`.

**Input:** `0 0 0 0 1 1 1`  
**Output:** `4`  
**Explanation:** Index `4` is the transition point where first `1` was observed.

**Input:** `arr[] = [1, 1, 1]`  
**Output:** `0`  
**Explanation:** There are no `0`s in the array, so the transition point is `0`.


```python
from typing import List

class Solution:
    def transitionPoint(self, arr: List[int]) -> int:
        for i, a in enumerate(arr):
            if a == 1:
                return i
        return -1
            
```

```python
# Test scaffold (from examples)

sol = Solution()

arr1 = [0, 0, 0, 1, 1]
print(sol.transitionPoint(arr1))
# Expected: 3

arr2 = [0, 0, 0, 0]
print(sol.transitionPoint(arr2))
# Expected: -1

arr3 = [0, 0, 0, 0, 1, 1, 1]
print(sol.transitionPoint(arr3))
# Expected: 4

arr4 = [1, 1, 1]
print(sol.transitionPoint(arr4))
# Expected: 0
```

# Pair with the given difference
**Last Updated:** 28 Mar, 2025

Given an **unsorted array** and an integer `x`, the task is to find if there exists a **pair of elements** in the array whose **absolute difference** is `x`.

## Examples

**Input:** `arr[] = [5, 20, 3, 2, 50, 80], x = 78`  
**Output:** `Yes`  
**Explanation:** The pair is `{2, 80}`.

**Input:** `arr[] = [90, 70, 20, 80, 50], x = 45`  
**Output:** `No`  
**Explanation:** No such pair exists.


```python
from typing import List

class Solution:
    def findPair(self, arr: List[int], x: int) -> bool:
        st = set()
    
        for num in arr:
            
            # Check if complement exists
            if (num + x) in st or (num - x) in st:
                return True
            
            st.add(num)
        
        return False
```

```python
# Test scaffold (from examples)

sol = Solution()

arr1 = [5, 20, 3, 2, 50, 80]
x1 = 78
print(sol.findPair(arr1, x1))
# Expected: True (Yes)

arr2 = [90, 70, 20, 80, 50]
x2 = 45
print(sol.findPair(arr2, x2))
# Expected: False (No)
```
