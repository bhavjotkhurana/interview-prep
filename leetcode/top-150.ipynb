# 80. Remove Duplicates from Sorted Array II

```python
class Solution(object):
    def removeDuplicates(self, nums):
        # k is the position where we write the next allowed element
        k = 0

        for num in nums:
            # If we've written fewer than 2 elements so far,
            # we always take the current one.
            # After that, we only take it if it's not the same as
            # the element at position k-2 (to avoid 3+ duplicates).
            if k < 2 or num != nums[k - 2]:
                nums[k] = num
                k += 1

        return k
```

```python
sol = Solution()

# Example 1
nums1 = [1, 1, 1, 2, 2, 3]
print("Before nums1:", nums1)
k1 = sol.removeDuplicates(nums1)
print("Returned k1:", k1)
print("After nums1:", nums1)
print("First k1 elements:", nums1[:k1])
print("-" * 40)

# Example 2
nums2 = [0, 0, 1, 1, 1, 1, 2, 3, 3]
print("Before nums2:", nums2)
k2 = sol.removeDuplicates(nums2)
print("Returned k2:", k2)
print("After nums2:", nums2)
print("First k2 elements:", nums2[:k2])
```

# 189. Rotate Array

```python
class Solution(object):
    def rotate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: None. Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        if n == 0:
            return
        
        k %= n          # make sure k is in [0, n-1]
        
        new = []
        a = k
        for i in range(n): 
            if a > 0: 
                new.append(nums[n - a])
                a -= 1
            else: 
                new.append(nums[i - k])
        
        nums[:] = new   # modify in-place
```

# 20. Valid Parentheses

```python
# Cell 2: Solution Skeleton
class Solution:
    def isValid(self, s):
        mapping = {')': '(', '}': '{', ']': '['}
        stack = []

        for char in s:
            if char in mapping.values():
                stack.append(char)
            elif char in mapping:
                if not stack or mapping[char] != stack.pop():
                    return False
        return not stack
```

# 35. Search Insert Position

```python
class Solution:
    def searchInsert(self, nums, target):
        low, high = 0, len(nums)
        while low < high: 
            mid = (low + high)//2 #floor division
            if target > nums[mid]: 
                low = mid + 1
            else: 
                high = mid
        return low
```

```python
# Cell 2: Solution Skeleton
class Solution(object):
    def plusOne(self, digits):
        rev = digits[::-1]

        for i in range(len(rev)):
            if rev[i] < 9:
                rev[i] += 1
                return rev[::-1]
            rev[i] = 0

        # If we exit the loop, it means all digits were 9
        rev.append(1)
        return rev[::-1]
```

```python
class Solution(object):
    def addBinary(self, a, b):
        i = len(a) - 1
        j = len(b) - 1
        carry = 0
        res = []

        while i >= 0 or j >= 0 or carry:
            bitA = int(a[i]) if i >= 0 else 0
            bitB = int(b[j]) if j >= 0 else 0

            total = bitA + bitB + carry
            res.append(str(total % 2))
            carry = total // 2

            i -= 1
            j -= 1

        return "".join(reversed(res))
```

# 69. Sqrt(x)

```python
# Cell 2: Solution Skeleton
class Solution:
    def mySqrt(self, x):
        square = 0 
        root = 0
        while square < x: 
            root += 1
            square = root * root
            if square > x: 
                return root - 1
        
        return root
```

```python
class Solution:
    def mySqrt(self, x):
        if x < 2:
            return x

        low, high = 1, x // 2
        ans = 0

        while low <= high:
            mid = (low + high) // 2
            sq = mid * mid

            if sq == x:
                return mid
            elif sq < x:
                ans = mid          # mid is a candidate (mid^2 <= x)
                low = mid + 1      # try bigger
            else:
                high = mid - 1     # mid^2 > x, try smaller

        return ans
```

# 83. Remove Duplicates from Sorted List

```python
# Cell 2: Solution Skeleton
from typing import Optional

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def deleteDuplicates(self, head):
        if not head:
            return head

        prev = head
        curr = head.next

        while curr:
            if curr.val == prev.val:
                prev.next = curr.next     # delete curr
                curr = curr.next          # only advance curr
            else:
                prev = curr               # now both move
                curr = curr.next

        return head


```

```python
# Cell 3: Test Scaffold
def build_linked_list(values):
    dummy = ListNode(0)
    cur = dummy
    for v in values:
        cur.next = ListNode(v)
        cur = cur.next
    return dummy.next

def linked_list_to_list(head):
    result = []
    cur = head
    while cur:
        result.append(cur.val)
        cur = cur.next
    return result

sol = Solution()

head1 = build_linked_list([1, 1, 2])
res1 = sol.deleteDuplicates(head1)
print(linked_list_to_list(res1))  # expected [1, 2]

head2 = build_linked_list([1, 1, 2, 3, 3])
res2 = sol.deleteDuplicates(head2)
print(linked_list_to_list(res2))  # expected [1, 2, 3]
```

# 55. Jump Game

```python
# Cell 2: Solution Skeleton
class Solution:
    def canJump(self, nums):
        max_index = {}
        n = len(nums)

        # store max jump from each index
        for i in range(n):
            max_index[i] = i + nums[i]

        # now track how far we can actually reach
        reach = 0
        for i in range(n):
            if i > reach:
                # we got to an index we can't even reach
                return False
            reach = max(reach, max_index[i])

        return reach >= n - 1
```

```python
# Cell 3: Test Scaffold

# Helper to build a binary tree from a level-order list (like LeetCode input)
from collections import deque

def build_tree(values):
    if not values or values[0] is None:
        return None
    
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    
    while queue and i < len(values):
        node = queue.popleft()
        
        if values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    
    return root

s = Solution()

print(s.maxDepth(build_tree([3,9,20,None,None,15,7])))  # Expected: 3
print(s.maxDepth(build_tree([1,None,2])))               # Expected: 2
```

# 110. Balanced Binary Tree

```python
# Cell 2: Solution Skeleton
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isBalanced(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        if not root: 
            return True
        
        left = root.left 
        right = root.right

        def maxDepth(root):
            if root is None: 
                return 0
            return max(maxDepth(root.left), maxDepth(root.right)) + 1
        
        left_depth = maxDepth(left)
        right_depth = maxDepth(right)

        diff = abs(left_depth - right_depth)

        return diff <= 1
    
```

```python
# Cell 3: Test Scaffold

from collections import deque

def build_tree(values):
    """Builds a binary tree from a level-order list (LeetCode-style)."""
    if not values or values[0] is None:
        return None
    
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    
    while queue and i < len(values):
        node = queue.popleft()
        
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    
    return root

s = Solution()

print(s.isBalanced(build_tree([3,9,20,None,None,15,7])))           # Expected: True
print(s.isBalanced(build_tree([1,2,2,3,3,None,None,4,4])))        # Expected: False
print(s.isBalanced(build_tree([])))                               # Expected: True
```

# 125. Valid Palindrome


```python
class Solution(object):
    def isPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        clean = ""
        for c in s: 
            if c.isalnum(): 
                clean += c.lower()
        
        start = 0 
        end = len(clean) - 1
        while start < end: 
            if clean[start] == clean[end]: 
                start += 1
                end -= 1
            else: 
                return False
        
        return True
```

# 168. Excel Sheet Column Title

```python
class Solution:
    def convertToTitle(self, columnNumber: int) -> str:
        result = []
        while columnNumber > 0:
            columnNumber -= 1                 # shift to 0-indexed
            remainder = columnNumber % 26     # find position (0-25)
            result.append(chr(65 + remainder))
            columnNumber //= 26               # reduce the number
        return "".join(result[::-1])
```
