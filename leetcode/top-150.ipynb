# 80. Remove Duplicates from Sorted Array II

```python
class Solution(object):
    def removeDuplicates(self, nums):
        # k is the position where we write the next allowed element
        k = 0

        for num in nums:
            # If we've written fewer than 2 elements so far,
            # we always take the current one.
            # After that, we only take it if it's not the same as
            # the element at position k-2 (to avoid 3+ duplicates).
            if k < 2 or num != nums[k - 2]:
                nums[k] = num
                k += 1

        return k
```

```python
sol = Solution()

# Example 1
nums1 = [1, 1, 1, 2, 2, 3]
print("Before nums1:", nums1)
k1 = sol.removeDuplicates(nums1)
print("Returned k1:", k1)
print("After nums1:", nums1)
print("First k1 elements:", nums1[:k1])
print("-" * 40)

# Example 2
nums2 = [0, 0, 1, 1, 1, 1, 2, 3, 3]
print("Before nums2:", nums2)
k2 = sol.removeDuplicates(nums2)
print("Returned k2:", k2)
print("After nums2:", nums2)
print("First k2 elements:", nums2[:k2])
```

# 189. Rotate Array

```python
class Solution(object):
    def rotate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: None. Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        if n == 0:
            return
        
        k %= n          # make sure k is in [0, n-1]
        
        new = []
        a = k
        for i in range(n): 
            if a > 0: 
                new.append(nums[n - a])
                a -= 1
            else: 
                new.append(nums[i - k])
        
        nums[:] = new   # modify in-place
```

# 20. Valid Parentheses

```python
# Cell 2: Solution Skeleton
class Solution:
    def isValid(self, s):
        mapping = {')': '(', '}': '{', ']': '['}
        stack = []

        for char in s:
            if char in mapping.values():
                stack.append(char)
            elif char in mapping:
                if not stack or mapping[char] != stack.pop():
                    return False
        return not stack
```

# 35. Search Insert Position

```python
class Solution:
    def searchInsert(self, nums, target):
        low, high = 0, len(nums)
        while low < high: 
            mid = (low + high)//2 #floor division
            if target > nums[mid]: 
                low = mid + 1
            else: 
                high = mid
        return low
```

# 66. Plus One

```python
# Cell 2: Solution Skeleton
class Solution(object):
    def plusOne(self, digits):
        rev = digits[::-1]

        for i in range(len(rev)):
            if rev[i] < 9:
                rev[i] += 1
                return rev[::-1]
            rev[i] = 0

        # If we exit the loop, it means all digits were 9
        rev.append(1)
        return rev[::-1]
```

```python
class Solution(object):
    def addBinary(self, a, b):
        i = len(a) - 1
        j = len(b) - 1
        carry = 0
        res = []

        while i >= 0 or j >= 0 or carry:
            bitA = int(a[i]) if i >= 0 else 0
            bitB = int(b[j]) if j >= 0 else 0

            total = bitA + bitB + carry
            res.append(str(total % 2))
            carry = total // 2

            i -= 1
            j -= 1

        return "".join(reversed(res))
```

# 69. Sqrt(x)

```python
# Cell 2: Solution Skeleton
class Solution:
    def mySqrt(self, x):
        square = 0 
        root = 0
        while square < x: 
            root += 1
            square = root * root
            if square > x: 
                return root - 1
        
        return root
```

```python
class Solution:
    def mySqrt(self, x):
        if x < 2:
            return x

        low, high = 1, x // 2
        ans = 0

        while low <= high:
            mid = (low + high) // 2
            sq = mid * mid

            if sq == x:
                return mid
            elif sq < x:
                ans = mid          # mid is a candidate (mid^2 <= x)
                low = mid + 1      # try bigger
            else:
                high = mid - 1     # mid^2 > x, try smaller

        return ans
```

# 83. Remove Duplicates from Sorted List

```python
# Cell 2: Solution Skeleton
from typing import Optional

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def deleteDuplicates(self, head):
        if not head:
            return head

        prev = head
        curr = head.next

        while curr:
            if curr.val == prev.val:
                prev.next = curr.next     # delete curr
                curr = curr.next          # only advance curr
            else:
                prev = curr               # now both move
                curr = curr.next

        return head


```

```python
# Cell 3: Test Scaffold
def build_linked_list(values):
    dummy = ListNode(0)
    cur = dummy
    for v in values:
        cur.next = ListNode(v)
        cur = cur.next
    return dummy.next

def linked_list_to_list(head):
    result = []
    cur = head
    while cur:
        result.append(cur.val)
        cur = cur.next
    return result

sol = Solution()

head1 = build_linked_list([1, 1, 2])
res1 = sol.deleteDuplicates(head1)
print(linked_list_to_list(res1))  # expected [1, 2]

head2 = build_linked_list([1, 1, 2, 3, 3])
res2 = sol.deleteDuplicates(head2)
print(linked_list_to_list(res2))  # expected [1, 2, 3]
```

# 55. Jump Game

```python
# Cell 2: Solution Skeleton
class Solution:
    def canJump(self, nums):
        max_index = {}
        n = len(nums)

        # store max jump from each index
        for i in range(n):
            max_index[i] = i + nums[i]

        # now track how far we can actually reach
        reach = 0
        for i in range(n):
            if i > reach:
                # we got to an index we can't even reach
                return False
            reach = max(reach, max_index[i])

        return reach >= n - 1
```

```python
# Cell 3: Test Scaffold
s = Solution()

print(s.canJump([2,3,1,1,4]))   # Expected: True
print(s.canJump([3,2,1,0,4]))   # Expected: False
print(s.canJump([0]))
```

```python
# Cell 3: Test Scaffold

# Helper to build a binary tree from a level-order list (like LeetCode input)
from collections import deque

def build_tree(values):
    if not values or values[0] is None:
        return None
    
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    
    while queue and i < len(values):
        node = queue.popleft()
        
        if values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    
    return root

s = Solution()

print(s.maxDepth(build_tree([3,9,20,None,None,15,7])))  # Expected: 3
print(s.maxDepth(build_tree([1,None,2])))               # Expected: 2
```

# 110. Balanced Binary Tree

```python
# Cell 2: Solution Skeleton
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isBalanced(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        if not root: 
            return True
        
        left = root.left 
        right = root.right

        def maxDepth(root):
            if root is None: 
                return 0
            return max(maxDepth(root.left), maxDepth(root.right)) + 1
        
        left_depth = maxDepth(left)
        right_depth = maxDepth(right)

        diff = abs(left_depth - right_depth)

        return diff <= 1
    
```

```python
# Cell 3: Test Scaffold

from collections import deque

def build_tree(values):
    """Builds a binary tree from a level-order list (LeetCode-style)."""
    if not values or values[0] is None:
        return None
    
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    
    while queue and i < len(values):
        node = queue.popleft()
        
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    
    return root

s = Solution()

print(s.isBalanced(build_tree([3,9,20,None,None,15,7])))           # Expected: True
print(s.isBalanced(build_tree([1,2,2,3,3,None,None,4,4])))        # Expected: False
print(s.isBalanced(build_tree([])))                               # Expected: True
```

# 125. Valid Palindrome


```python
class Solution(object):
    def isPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        clean = ""
        for c in s: 
            if c.isalnum(): 
                clean += c.lower()
        
        start = 0 
        end = len(clean) - 1
        while start < end: 
            if clean[start] == clean[end]: 
                start += 1
                end -= 1
            else: 
                return False
        
        return True
```

# 168. Excel Sheet Column Title

```python
class Solution:
    def convertToTitle(self, columnNumber: int) -> str:
        result = []
        while columnNumber > 0:
            columnNumber -= 1                 # shift to 0-indexed
            remainder = columnNumber % 26     # find position (0-25)
            result.append(chr(65 + remainder))
            columnNumber //= 26               # reduce the number
        return "".join(result[::-1])
```

```python
from typing import List

class Solution:
    def summaryRanges(self, nums: List[int]) -> List[str]:
        i = 0 
        ranges = []
        while i < len(nums): 
            start = nums[i]
            while i + 1 < len(nums) and nums[i] + 1 == nums[i + 1]: 
                i += 1
            
            if start != nums[i]: 
                ranges.append(f"{start}" + "->" + f"{nums[i]}")
            else:
                ranges.append(f"{nums[i]}")
                
            i += 1
        return ranges 
```

# 243. Shortest Word Distance

```python
from typing import List

class Solution:
    def shortestDistance(self, wordsDict: List[str], word1: str, word2: str) -> int:
        distance = 3 * 10**4  # use exponentiation
        loc1 = -1
        loc2 = -1
        
        for i in range(len(wordsDict)):
            if wordsDict[i] == word1:
                loc1 = i
            if wordsDict[i] == word2:
                loc2 = i
            
            # Only compute distance once both positions are known
            if loc1 != -1 and loc2 != -1:
                distance = min(distance, abs(loc1 - loc2))
        
        return distance
    
```

# 252. Meeting Rooms


```python
from typing import List

class Solution:
    def canAttendMeetings(self, intervals: List[List[int]]) -> bool:
        # Sort intervals by start time (required for proper overlap checking)
        if not intervals: 
            return True
        intervals.sort(key=lambda x: x[0])
        prev_end = intervals[0][1]
        for i in range(1, len(intervals)): 
            start, end = intervals[i]
            if start < prev_end: 
                return False
            prev_end = end
        return True
```

# 455. Assign Cookies

```python
from typing import List

class Solution:
    def findContentChildren(self, g: List[int], s: List[int]) -> int:
        if not g or not s: 
            return 0
        
        g.sort()
        s.sort()

        gn = len(g)
        sn = len(s)

        count = 0
        i = 0  # cookies
        j = 0  # children

        while i < sn and j < gn:
            if s[i] >= g[j]:
                count += 1
                i += 1
                j += 1
            else:  # s[i] < g[j], cookie too small -> try next cookie
                i += 1

        return count
```

# 463. Island Perimeter


```python
from typing import List

class Solution:
    def islandPerimeter(self, grid: List[List[int]]) -> int:
        perimeter = 0
        rows = len(grid)
        cols = len(grid[0])
        for i in range(rows): 
            for j in range(cols): 
                if grid[i][j] == 1:
                    perimeter += 4
                    if i - 1 >= 0 and grid[i - 1][j] == 1: #up
                        perimeter -= 1
                    if i + 1 < rows and grid[i + 1][j] == 1: #down
                        perimeter -= 1
                    if j - 1 >= 0 and grid[i][j - 1] == 1: #left
                        perimeter -= 1
                    if j + 1 < cols and grid[i][j + 1] == 1: #right
                        perimeter -= 1
        return perimeter
```

```python
from typing import List

class Solution:
    def arrayPairSum(self, nums: List[int]) -> int:
        nums.sort()
        i = 0
        sum = 0
        while i < len(nums): 
            sum += min(nums[i], nums[i + 1])
            i += 2
        return sum
        
```

# 566. Reshape the Matrix

```python
from typing import List

class Solution:
    def matrixReshape(self, mat: List[List[int]], r: int, c: int) -> List[List[int]]:
        m = len(mat)
        n = len(mat[0])
        if r*c != m*n:
            return mat
        new = [[0 for _ in range(c)] for _ in range(r)]
        flat = []
        for row in mat:
            flat.extend(row)
        a = 0
        for i in range(r): 
            for j in range(c): 
                new[i][j] = flat[a]
                a += 1
        return new
```

# 594. Longest Harmonious Subsequence

```python
from typing import List

class Solution:
    def findLHS(self, nums: List[int]) -> int:
        dt = {}
        for i in range(len(nums)): 
            if nums[i] in dt: 
                dt[nums[i]] += 1
            else: 
                dt[nums[i]] = 1
        ans = 0
        for x in dt: 
            if x + 1 in dt: 
                ans = max(ans, dt[x] + dt[x + 1])
        return ans
```

# 605. Can Place Flowers


```python
from typing import List

class Solution:
    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
        count = 0
        for i in range(len(flowerbed)):
            if flowerbed[i] == 0:
                empty_left = (i == 0) or (flowerbed[i - 1] == 0)
                empty_right = (i == len(flowerbed) - 1) or (flowerbed[i + 1] == 0)
                if empty_left and empty_right:
                    flowerbed[i] = 1  # plant flower
                    count += 1

        return count >= n
```

# 643. Maximum Average Subarray I

```python
from typing import List

class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        n = len(nums)
        if n < k:
            return 0
    
    # Initialize: sum the first window
        window_sum = sum(nums[:k])      # This costs O(k) once
        max_sum = window_sum
    
    # Slide: update with delta only
        for i in range(k, n):
            window_sum -= nums[i - k]   # Drop the tail (Blue flash)
            window_sum += nums[i]       # Grab the head (Green flash)
            max_sum = max(max_sum, window_sum)  # Snapshot the score
    
        return max_sum/k
```

# 645. Set Mismatch

```python
from typing import List

class Solution:
    def findErrorNums(self, nums: List[int]) -> List[int]:
        #find the duplicate
        nums.sort()
        dup = nums[0]
        for i in range(1, len(nums)): 
            if dup == nums[i]: 
                break
            dup = nums[i]

        for i in range(1, len(nums) + 1): 
            if nums[i - 1] != i: 
                missing = i
                break
            if nums[i - 1] == dup: 
                i += 1
        
        return [dup, missing]
            

```

```python
# Additional test cases for findErrorNums

s = Solution()

# Example 1
nums1 = [1, 2, 2, 4]
print("Example 1:", s.findErrorNums(nums1))  # Expected: [2, 3]

# Example 2
nums2 = [1, 1]
print("Example 2:", s.findErrorNums(nums2))  # Expected: [1, 2]


# Additional Tests

# 1. Missing last number
nums3 = [1, 2, 3, 4, 4]
print("Test 1 (missing 5):", s.findErrorNums(nums3))  # Expected: [4, 5]

# 2. Missing first number
nums4 = [2, 2]
print("Test 2 (missing 1):", s.findErrorNums(nums4))  # Expected: [2, 1]

# 3. Duplicate in the middle, missing another
nums5 = [3, 1, 2, 5, 3]
print("Test 3 (dup 3, missing 4):", s.findErrorNums(nums5))  # Expected: [3, 4]

# 4. Longer array, duplicate near end
nums6 = [1, 2, 3, 4, 5, 6, 7, 7]
print("Test 4 (missing 8):", s.findErrorNums(nums6))  # Expected: [7, 8]

# 5. Duplicate at start
nums7 = [2, 1, 2]
print("Test 5 (missing 3):", s.findErrorNums(nums7))  # Expected: [2, 3]

# 6. Random small case
nums8 = [4, 3, 6, 2, 1, 1]
print("Test 6 (dup 1, missing 5):", s.findErrorNums(nums8))  # Expected: [1, 5]

# 7. Larger random-like case
nums9 = [1, 5, 3, 2, 2, 6, 7, 8, 9, 10]
print("Test 7 (missing 4):", s.findErrorNums(nums9))  # Expected: [2, 4]
```

# 661. Image Smoother


```python
from typing import List

class Solution:
    def imageSmoother(self, img: List[List[int]]) -> List[List[int]]:
        row = len(img)
        col = len(img[0])
        smooth = [[0 for _ in range(col)] for _ in range(row)]
        total = 0
        nbr = 0
        for i in range(row): 
            for j in range(col): 
                total += img[i][j]
                nbr += 1
                #up
                if i > 0: 
                    total += img[i - 1][j]
                    nbr += 1
                #down
                if i < row - 1: 
                    total += img[i + 1][j]
                    nbr += 1
                #left
                if j > 0: 
                    total += img[i][j - 1]
                    nbr += 1
                #right
                if j < col - 1: 
                    total += img[i][j + 1]
                    nbr += 1
                #up-left
                if i > 0 and j > 0: 
                    total += img[i - 1][j - 1]
                    nbr += 1
                #up-right
                if i > 0 and j < col - 1:
                    total += img[i - 1][j + 1]
                    nbr += 1
                #down-left
                if i < row - 1 and j > 0:
                    total += img[i + 1][j - 1]
                    nbr += 1
                #down-right
                if i < row - 1 and j < col - 1: 
                    total += img[i + 1][j + 1]
                    nbr += 1
                avg = total // nbr
                smooth[i][j] = avg
                total = 0 
                nbr = 0
        return smooth
```

```python
# Test scaffold for the examples

s = Solution()

# Example 1
img1 = [[1,1,1],[1,0,1],[1,1,1]]
print(s.imageSmoother(img1))  # Expected: [[0,0,0],[0,0,0],[0,0,0]]

# Example 2
img2 = [[100,200,100],[200,50,200],[100,200,100]]
print(s.imageSmoother(img2))  
# Expected: [[137,141,137],[141,138,141],[137,141,137]]
```

# 674. Longest Continuous Increasing Subsequence


```python
from typing import List

class Solution:
    def findLengthOfLCIS(self, nums: List[int]) -> int:
        if not nums: 
            return 0
        seq = 1
        longest = 1
        for i in range(1, len(nums)): 
            if nums[i] > nums[i - 1]: 
                seq += 1
            else: 
                longest = max(longest, seq)
                seq = 1
        return max(longest, seq) #needed one final check
```

```python
# Test scaffold for 674. Longest Continuous Increasing Subsequence

s = Solution()

# Example 1
nums1 = [1, 3, 5, 4, 7]
print(s.findLengthOfLCIS(nums1))  # Expected output: 3

# Example 2
nums2 = [2, 2, 2, 2, 2]
print(s.findLengthOfLCIS(nums2))  # Expected output: 1
```

# 682. Baseball Game

You are keeping the scores for a baseball game with strange rules. At the beginning of the game, you start with an empty record.

You are given a list of strings `operations`, where `operations[i]` is the ith operation you must apply to the record and is one of the following:

- **An integer `x`**  
  Record a new score of `x`.

- **"+"**  
  Record a new score that is the sum of the previous two scores.

- **"D"**  
  Record a new score that is the double of the previous score.

- **"C"**  
  Invalidate the previous score, removing it from the record.

Return the sum of all the scores on the record after applying all the operations.

The test cases are generated such that the answer and all intermediate calculations fit in a 32-bit integer and that all operations are valid.



```python
from typing import List

class Solution:
    def calPoints(self, o: List[str]) -> int:
        points = []
        n = len(o)
        for i in range(n): 
            if o[i] == '+': 
                points.append(points[-1] + points[-2])
            elif o[i] == 'D': 
                points.append(points[-1] * 2)
            elif o[i] == 'C': 
                points.pop()
            else: 
                points.append(int(o[i]))
        return sum(points)
```

```python
# Test scaffold for 682. Baseball Game

s = Solution()

# Example 1
ops1 = ["5", "2", "C", "D", "+"]
print(s.calPoints(ops1))  # Expected output: 30

# Example 2
ops2 = ["5", "-2", "4", "C", "D", "9", "+", "+"]
print(s.calPoints(ops2))  # Expected output: 27

# Example 3
ops3 = ["1", "C"]
print(s.calPoints(ops3))  # Expected output: 0
```
