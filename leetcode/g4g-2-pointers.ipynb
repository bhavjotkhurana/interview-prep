# Remove All Occurrences of an Element in an Array

Given an integer array `arr[]` and an integer `ele`, the task is to remove all occurrences of `ele` from `arr[]` **in-place** and return the number of elements which are not equal to `ele`.

If there are `k` number of elements which are not equal to `ele`, then the input array `arr[]` should be modified such that the first `k` elements contain the elements which are not equal to `ele`, and then the remaining elements can contain any value.

**Note:** The order of the first `k` elements may be changed.

---

## Examples

### Example 1
**Input:**  
`arr[] = [3, 2, 2, 3], ele = 3`  

**Output:**  
`2`  

**Explanation:**  
The answer is 2 because there are 2 elements which are not equal to 3, and `arr[]` will be modified such that the first 2 elements contain the elements which are not equal to 3 and remaining elements can contain any element.  
So, modified `arr[] = [2, 2, _, _]`

---

### Example 2
**Input:**  
`arr[] = [0, 1, 3, 0, 2, 2, 4, 2], ele = 2`  

**Output:**  
`5`  

**Explanation:**  
The answer is 5 because there are 5 elements which are not equal to 2, and `arr[]` will be modified such that the first 5 elements contain the elements which are not equal to 2 and remaining elements can contain any element.  
So, modified `arr[] = [0, 1, 3, 0, 4, _, _, _]`


```python
from typing import List

class Solution:
    def removeElement(self, arr: List[int], ele: int) -> int:
        idx = 0 
        for i in range(len(arr)): 
            if arr[i] != ele: 
                arr[idx] = arr[i]
                idx += 1
        return idx
```

```python
# Test scaffold

sol = Solution()

arr1 = [3, 2, 2, 3]
k1 = sol.removeElement(arr1, 3)
print(k1, arr1)
# Expected output:
# 2
# First 2 elements should be [2, 2] in any order

arr2 = [0, 1, 3, 0, 2, 2, 4, 2]
k2 = sol.removeElement(arr2, 2)
print(k2, arr2)
# Expected output:
# 5
# First 5 elements should be [0, 1, 3, 0, 4] in any order
```

# Move all Zeros to End of Array

Given an array of integers `arr[]`, move all the zeros to the end of the array while maintaining the relative order of all non-zero elements.

---

## Examples

### Example 1
**Input:**  
`arr[] = [1, 2, 0, 4, 3, 0, 5, 0]`  

**Output:**  
`[1, 2, 4, 3, 5, 0, 0, 0]`  

**Explanation:**  
There are three 0s that are moved to the end.

---

### Example 2
**Input:**  
`arr[] = [10, 20, 30]`  

**Output:**  
`[10, 20, 30]`  

**Explanation:**  
No change in array as there are no 0s.

---

### Example 3
**Input:**  
`arr[] = [0, 0]`  

**Output:**  
`[0, 0]`  

**Explanation:**  
No change in array as there are all 0s.


```python
from typing import List

class Solution:
    def moveZeroes(self, arr: List[int]) -> None:
        idx = 0
        for i in range(len(arr)): 
            if arr[i] != 0: 
                arr[idx] = arr[i]
                idx += 1

        for i in range(idx, len(arr)): 
            arr[i] = 0
        
```

```python
# Test scaffold

sol = Solution()

arr1 = [1, 2, 0, 4, 3, 0, 5, 0]
sol.moveZeroes(arr1)
print(arr1)
# Expected output: [1, 2, 4, 3, 5, 0, 0, 0]

arr2 = [10, 20, 30]
sol.moveZeroes(arr2)
print(arr2)
# Expected output: [10, 20, 30]

arr3 = [0, 0]
sol.moveZeroes(arr3)
print(arr3)
# Expected output: [0, 0]
```

# Remove duplicates from Sorted Array

Given a sorted array `arr[]` of size `n`, the goal is to rearrange the array so that all distinct elements appear at the beginning in sorted order. Additionally, return the length of this distinct sorted subarray.

**Note:** The elements after the distinct ones can be in any order and hold any value, as they don't affect the result.

---

## Examples

### Example 1
**Input:**  
`arr[] = [2, 2, 2, 2, 2]`  

**Output:**  
`[2]`  

**Explanation:**  
All the elements are 2, so only keep one instance of 2.

---

### Example 2
**Input:**  
`arr[] = [1, 2, 2, 3, 4, 4, 4, 5, 5]`  

**Output:**  
`[1, 2, 3, 4, 5]`

---

### Example 3
**Input:**  
`arr[] = [1, 2, 3]`  

**Output:**  
`[1, 2, 3]`  

**Explanation:**  
No change as all elements are distinct.


```python
from typing import List

class Solution:
    def removeDuplicates(self, arr: List[int]) -> int:
        if not arr: 
            return 0
        idx = 1
        for i in range(1, len(arr)): 
            if arr[i - 1] != arr[i]: 
                arr[idx] = arr[i]
                idx += 1
        return idx
        
```

```python
# Test scaffold

sol = Solution()

arr1 = [2, 2, 2, 2, 2]
k1 = sol.removeDuplicates(arr1)
print(k1, arr1[:k1])
# Expected output:
# 1 [2]

arr2 = [1, 2, 2, 3, 4, 4, 4, 5, 5]
k2 = sol.removeDuplicates(arr2)
print(k2, arr2[:k2])
# Expected output:
# 5 [1, 2, 3, 4, 5]

arr3 = [1, 2, 3]
k3 = sol.removeDuplicates(arr3)
print(k3, arr3[:k3])
# Expected output:
# 3 [1, 2, 3]
```
